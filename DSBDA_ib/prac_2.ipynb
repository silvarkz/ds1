```python
#outliers:

from scipy import stats #scipy library aahe aani stats tyatla module aahe 
```

```python
z=np.abs(stats.zscore(df['Math_Score'])) 
#abs=absolute (non_negative)
```

```python
print(z)
```

```python
threshold=0.18   
```

```python
#Detecting Outliers:
sample_outliers=np.where(z<threshold)
sample_outliers
```

```python
#removing outliers:

new_df=df
for i in sample_outliers:
    new_df.drop(i,inplace=True)
new_df
```
